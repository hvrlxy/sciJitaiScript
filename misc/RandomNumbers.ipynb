# Import libraries

```python
from numpy.random import seed
from numpy.random import randint
from numpy.random import permutation
import numpy as np
import pandas as pd
# seed random number generator
seed(1)
```

# Random permutation of numbers for 6 days a week for n number of weeks

```python
# Change number of weeks here
num_weeks = 3

# Creating random permutation of numbers for 6 days a week for num_weeks
weeks = list()
for i in range(num_weeks):
    arr = np.arange(6)
    values = list(np.random.permutation(arr))
    weeks.append(values)
weeks
```

# Adding Week and Day Columns

```python
weeks1 = pd.DataFrame(weeks)
weeks1['Week'] = weeks1.index.values + 1
weeks1 = weeks1.rename(columns={0: "Day 1", 1: 'Day 2', 2: "Day 3", 3: "Day 4", 4: "Day 5", 5: 'Day 6'})
weeks2 = pd.melt(weeks1, id_vars=['Week'], value_vars=["Day 1", "Day 2", "Day 3", "Day 4", "Day 5", "Day 6"], ignore_index=False,
        var_name='Day', value_name='Beginning of Day - Goal Setting')
weeks3 = weeks2
weeks3
```

# Replacing random numbers with goal names

```python
weeks3["Beginning of Day - Goal Setting"] = weeks3["Beginning of Day - Goal Setting"].replace([0, 1], "Standard Goal")
weeks3["Beginning of Day - Goal Setting"] = weeks3["Beginning of Day - Goal Setting"].replace([2, 3], "Tailored Goal")
weeks3["Beginning of Day - Goal Setting"] = weeks3["Beginning of Day - Goal Setting"].replace([4, 5], "No Goal")
weeks3
```

# Getting random integers depending on goal type and creating messages

```python
weeks3['First Message'] = 'na'
weeks3['Second Message'] = 'na'
for i in range(len(weeks3)):
    # Standard goal three options for first and second messages
    if weeks3["Beginning of Day - Goal Setting"].iloc[i] == "Standard Goal": 
        # Get random integer between 0 - 2 
        goal1 = randint(0, 3, 1)
        if goal1 == 0:
            goal1 = "Jit Message: To-Date: Min Achieved"
        if goal1 == 1:
            goal1 = "Jit Message: To-Do: Remaining"
        if goal1 == 2: 
            goal1 = "No Message"
        weeks3['First Message'].iloc[i] = str(goal1)  
        # Get random integer between 0 - 2
        goal2 = randint(0, 3, 1)
        if goal2 == 0:
            goal2 = "Jit Message: To-Date: Min Achieved"
        if goal2 == 1:
            goal2 = "Jit Message: To-Do: Remaining"
        if goal2 == 2: 
            goal2 = "No Message"
        weeks3['Second Message'].iloc[i] = str(goal2)
    if weeks3["Beginning of Day - Goal Setting"].iloc[i] == "Tailored Goal":
        # Get random integer between 0 - 1
        goal1 = randint(0, 3, 1)
        if goal1 == 0:
            goal1 = "Jit Message: To-Date: Min Achieved"
        if goal1 == 1:
            goal1 = "Jit Message: To-Do: Remaining"
        if goal1 == 2: 
            goal1 = "No Message"
        weeks3['First Message'].iloc[i] = str(goal1)  
        goal2 = randint(0, 3, 1)
        if goal2 == 0:
            goal2 = "Jit Message: To-Date: Min Achieved"
        if goal2 == 1:
            goal2 = "Jit Message: To-Do: Remaining"
        if goal2 == 2: 
            goal2 = "No Message"
        weeks3['Second Message'].iloc[i] = str(goal2)
    if weeks3["Beginning of Day - Goal Setting"].iloc[i] == "No Goal":
        nogoal1 = randint(0, 2, 1)
        if nogoal1 == 1:
            nogoal1 = "Jit Message - Min Achieved"
        if nogoal1 == 0: 
            nogoal1 = "No Message"
        weeks3['First Message'].iloc[i] = str(nogoal1)
        nogoal2 = randint(0, 2, 1)
        if nogoal2 == 1:
            nogoal2 = "Jit Message - Min Achieved"
        if nogoal2 == 0: 
            nogoal2 = "No Message"
        weeks3['Second Message'].iloc[i] = str(nogoal2)
weeks3
```

# Modifying index

```python
weeks4 = weeks3.set_index(["Week", "Day"])
weeks4 = weeks4.groupby(level=["Week","Day"]).sum()
weeks4
```

# Save to CSV

```python
# Save file to directory as .csv
weeks4.to_csv("RandomNumbers.csv")
```
